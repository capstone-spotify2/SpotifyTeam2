```python
## Add up the total occurrence for each word in each type
total_happy = full_happy.iloc[:,2:].sum()
total_sad = full_sad.iloc[:,2:].sum()
total_energetic = full_energetic.iloc[:,2:].sum()
total_relax = full_relax.iloc[:,2:].sum()
```

```python
## Sort by total number of occurences
sorted_happy = total_happy.sort_values(ascending=False, inplace=False)
sorted_sad = total_sad.sort_values(ascending=False, inplace=False)
sorted_energetic = total_energetic.sort_values(ascending=False, inplace=False)
sorted_relax = total_relax.sort_values(ascending=False, inplace=False)
```

```python
## Find the intersection of top 30 occurences in each category
## Try using their intersection to be the stop list - words that do not add much meaning/value
inter1 = np.intersect1d(sorted_happy.index.values[:30],sorted_sad.index.values[:30])
inter2 = np.intersect1d(sorted_energetic.index.values[:30],sorted_relax.index.values[:30])
stop_list = np.intersect1d(inter1, inter2)
stop_list
```

```python
## Try using Python package for stop words
from stop_words import get_stop_words
stop_words = get_stop_words('en')
stop_words
```

```python
## Get the normalized top 100 sad words after removing stop words
## Normalize with the count of word w devided by max count across all words
sorted_happy_df = sorted_happy.to_frame()
sorted_happy_df.reset_index(level=0, inplace=True)
happy_final = sorted_happy_df.loc[~sorted_happy_df['index'].isin(stop_words)]
happy_final = happy_final.iloc[:100]
happy_final.iloc[:,1] /= happy_final.iloc[0,1]
happy_final.columns = ['word','count']
happy_final.head(5)
```

```python
## Get the normalized top 100 sad words after removing stop words
## Normalize with the count of word w devided by max count across all words
sorted_sad_df = sorted_sad.to_frame()
sorted_sad_df.reset_index(level=0, inplace=True)
sad_final = sorted_sad_df.loc[~sorted_sad_df['index'].isin(stop_words)]
sad_final = sad_final.iloc[:100]
sad_final.iloc[:,1] /= sad_final.iloc[0,1]
sad_final.columns = ['word','count']
sad_final.head(5)
```
